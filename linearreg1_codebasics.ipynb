```python
import pandas as pd
import sklearn
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/codebasics/py/master/ML/1_linear_reg/Exercise/canada_per_capita_income.csv")
```

```python
#df.head()
df
```

```python
from sklearn import linear_model
import numpy as np
import matplotlib.pyplot as plt
```

```python
df.columns
```

```python
plt.xlabel("year")
plt.ylabel("per capita income")
plt.plot(df['year'],df['per capita income (US$)'])
plt.show()
```

```python
reg=linear_model.LinearRegression()
```

```python
reg.fit(df[['year']],df['per capita income (US$)'])
```

```python
reg.predict([[2020]])
```

So,Canada's per capita income for 2020 is 41288.69409442

```python
import pickle
```

This module will help us to save our trained model into a file so that we can share this file to others/cilent.Then the client
can just load the file, give the required input alone and get the predicted value instead of trainingthe model each and every time.

```python
with open("model_pickle",'wb')as f:
     pickle.dump(reg,f)
```

We have created a file named model_pickle(created a memory to be precise) and write the binary data and dumped the trained model('reg') into it.    

```python
with open("model_pickle",'rb')as f:
    modelfile=pickle.load(f)
```

In the above step we creates a memory named model_pickle and dumped the trained model into it and 
now we have created a 
object called modelfile to that memory to call it every time and
get the predicted value to avoid traing the model each and every time.

```python
modelfile.predict([[2020]])
```

see we got the same as "Out [41]".....and also check the model_pickle filecreated in the directory 
which is nothing but  our saved model....as a file.

Just like pickle api,there is also sklearn joblib...which perform same function...but in joblib we can directly give the file name..check the below code.

```python
import joblib 
```

```python
joblib.dump(reg,'model_joblib')
modelfilejoblib=joblib.load('model_joblib')
modelfilejoblib.predict([[2020]])
```

tadaaaah!!!!!! but the difference between pickle and joblib is if your model has more numpy arrays to deal with..then it is believed that joblib can be preffered to pickle.
